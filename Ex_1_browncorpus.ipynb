```python
import nltk
from nltk.tag import hmm
from nltk.probability import LidstoneProbDist
from nltk.corpus import brown
```

```python
nltk.download('brown')
nltk.download('universal_tagset')
```

```python
print("Sample Original document:", brown.raw()[:500])
```

```python
tagged_sentences = brown.tagged_sents(tagset='universal')
print("Total Sentences:", len(tagged_sentences))
print("Sample Sentence:", tagged_sentences[0][:10])
```

```python
train_size = int(0.8 * len(tagged_sentences))
train_data = tagged_sentences[:train_size]
test_data = tagged_sentences[train_size:]

print("Training Sentences:", len(train_data))
print("Testing Sentences:", len(test_data))
```

```python
train = hmm.HiddenMarkovModelTrainer()
model = train.train_supervised(train_data, estimator=lambda fd, bins: LidstoneProbDist(fd, 0.1, bins))
```

```python
sentence = "The quick brown fox jumps over the lazy dog".lower().split()
tagged = model.tag(sentence)

print("Tagged Sentence:", tagged)
```

```python
acc = model.accuracy(test_data)
print("Model Accuracy:", acc)
```

```python
print("\nTop transitions from DET:")
for tag in model._transitions['DET'].samples():
    print(tag, model._transitions['DET'].prob(tag))
```

```python
tag = 'NOUN'
print(f"Top emissions for tag {tag}:")
for word in sorted(model._outputs[tag].samples(), key=lambda w: model._outputs[tag].prob(w), reverse=True)[:10]:
    print(word, round(model._outputs[tag].prob(word), 4))
```
